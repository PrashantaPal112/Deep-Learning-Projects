```python
def text_to_indices(sentence,vocab):
  numerical_sentence=[]
  for token in sentence:
    if token in vocab:
      numerical_sentence.append(vocab[token])
    else:
      numerical_sentence.append(vocab['<unk>'])
  return numerical_sentence
```

```python
input_numeric_sentence=[]
for sentence in input_sentence:
  input_numeric_sentence.append(text_to_indices(word_tokenize(sentence.lower()),vocab))
input_numeric_sentence
```

```python
class CustomDataset(Dataset):
  def __init__(self,X,y):
    self.X=X
    self.y=y
  def __len__(self):
    return len(self.X)
  def __getitem__(self,idx):
    return self.X[idx],self.y[idx]
```

# LSTM Cell

```python
class LSTMmodel(nn.Module):
  def __init__(self,vocab_size):
    super().__init__()
    self.embadding=nn.Embedding(vocab_size,100)
    self.lstm=nn.LSTM(100,150,batch_first=True)
  def forward(self,x):
    embedded=self.embadding(x)
    intermidiate_hidden_state,(final_hidden_state,final_cell_state)=self.lstm(embedded)
    output=self.fc(final_hidden_state.unsqueeze(0))
    return output
```

```python
model=LSTMmodel(len(vocab))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

# Training Loop

```python
epochs = 50
learning_rate = 0.001

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)


# training loop

for epoch in range(epochs):
  total_loss = 0

  for batch_x, batch_y in dataloader:

    batch_x, batch_y = batch_x.to(device), batch_y.to(device)

    optimizer.zero_grad()

    output = model(batch_x)

    loss = criterion(output, batch_y)

    loss.backward()

    optimizer.step()

    total_loss = total_loss + loss.item()

  print(f"Epoch: {epoch + 1}, Loss: {total_loss:.4f}")
```

# Pediction Function

```python
def prediction(model,vocab,text):

  #tokenize
  tokenized_text = word_tokenize(text.lower())


  #numerical indices
  numerical_text = text_to_indices(tokenized_text,vocab)

  #padding
  padded_text = torch.tensor([0]* (25-len(numerical_text)) + numerical_text , dtype=torch.long).unsqueeze(0)


  #send to model
  output = model(padded_text)

  #predicted index

  value, index = torch.max(output,dim=1)
  #merge with text

  return text + " " + list(vocab.keys())[index]
```
